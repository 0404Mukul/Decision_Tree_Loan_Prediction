```python
import pandas as pd 
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv(r'loan_approval_dataset.csv')
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
# Droping the columns
```

```python
data.drop(columns = ['loan_id', ' no_of_dependents', ' education', ' self_employed'],inplace = True)
```

```python
data.columns
```

## Converting Loan_Statue columns into 0 and 1
0 --> Reject
1 --> Approve

```python
data[' loan_status']=data[' loan_status'].apply(lambda x: 1 if x == ' Approved' else 0)
```

```python
data[' loan_status']
```

```python
## Splitting the Data into Training and Testing
```

```python
x=data.drop(columns =' loan_status')
y=data[' loan_status']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
## Implementation of StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train= scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = RandomForestClassifier(random_state=2)
```

```python
model.fit(x_train , y_train)
```

```python
Y_train_pred = model.predict(x_train)
print("Accuracy: ", accuracy_score(y_train,Y_train_pred))
```

```python
print("Accuracy:" ,accuracy_score(y_train,Y_train_pred))
print("Classification Report: " ,classification_report(y_train,Y_train_pred))
print("Confusion Matrix: " , confusion_matrix(y_train,Y_train_pred))
```

```python
X_test_pred = model.predict(x_test)
print("Accuracy: ", accuracy_score(y_test,X_test_pred))
```

```python
param_grid = {
    'n_estimators':[100,200,300],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[2,4,6],
    'bootstrap':[True ,False]
}
```

```python
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=2),
                         param_grid = param_grid,
                         cv=5,
                         n_jobs=1,
                         verbose=2,
                         scoring='accuracy')
grid_search.fit(x_train,y_train)
```

```python
print("Best Parameters:" ,grid_search.best_params_)

best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(x_test)

print("Accuracy:" ,accuracy_score(y_test,y_pred_best))
print("Classification Report: " ,classification_report(y_test,y_pred_best))
print("Confusion Matrix: " , confusion_matrix(y_test,y_pred_best))
```

```python

# Apply best parameters to new model
best_model = RandomForestClassifier(
    bootstrap=False,
    max_depth=None,
    min_samples_leaf=2,
    min_samples_split=5,
    n_estimators=200,
    random_state=2
)
```

```python
# Fit the model on training data
best_model.fit(x_train, y_train)
```

```python
# Predict on test data
y_pred_best1 = best_model.predict(x_test)
```

```python
print(" Accuracy:", accuracy_score(y_test, y_pred_best1))
print("Classification Report:\n", classification_report(y_test, y_pred_best1))
print(" Confusion Matrix:\n", confusion_matrix(y_test, y_pred_best1))
```
